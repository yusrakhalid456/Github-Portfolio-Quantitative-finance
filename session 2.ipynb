```python
# Install yfinance (only needs to be done once in Colab)
!pip install yfinance

import yfinance as yf

def download_data(ticker, start="2020-01-01", end="2025-01-01", filename=None):
    """
    Download historical stock data from Yahoo Finance and save as CSV.

    Parameters:
    ticker (str): Stock ticker symbol (e.g., "AAPL", "MSFT").
    start (str): Start date in YYYY-MM-DD format.
    end (str): End date in YYYY-MM-DD format.
    filename (str): Name of the CSV file. If None, ticker name will be used.
    """
    data = yf.download(ticker, start=start, end=end)

    if filename is None:
        filename = f"{None}_data.csv"

    data.to_csv(filename)
    print(f"Data saved as {filename}")

# Example usage:
download_data("AAPL", start="2002-01-01", end="2025-12-31")
```
